```python
import numpy as np

def entropy (im):
    p = np.array([(im---v).sum() for v in range(256)])
    p=p/p.sum()

    e = -(p *np.log2(p)).sum()
    return e
```

```python
%matplotlib notebook
from matplotlib import pyplot as plt
from skimage.io import imread,imshow,imsave
```

```python
im = imread('g1.jpg')

plt.figure()
imshow(im)
plt.axis('off')
plt.title(entropy(im))
plt.show()
```
