```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('selena1.jpg')

plt.figure(figsize=(12,6))
plt.imshow(img)
```

```python
image_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12,6))
plt.imshow(img)
```

```python
blurred_image = cv2.blur(image_RGB , (100,100))
```

```python
plt.figure(figsize=(12,6))
plt.imshow(blurred_image)
```

```python
gaussian_blur = cv2.GaussianBlur(image_RGB , (101,101) , 0 )

plt.figure(figsize=(12,6))
plt.imshow(gaussian_blur)
```

```python
bilatera_blur = cv2.bilateralFilter(image_RGB , 23,51,51)

plt.figure(figsize=(12,6))
plt.imshow(bilatera_blur)
```

```python
original_noise = cv2.Canny (image_RGB, 2 , 15)

plt.figure(figsize=(12,6))
plt.imshow(original_noise)
```

```python
bilateral_noise = cv2.Canny (bilatera_blur, 2 , 15)

plt.figure(figsize=(12,6))
plt.imshow(bilateral_noise)
```
